```python
class Pelicula:
    def __init__(self, titulo="", director="", anio=0, genero="", duracion=0.0, calificacion=0.0, idioma="", estado=""):
        self.titulo = titulo
        self.director = director
        self.anio = anio
        self.genero = genero
        self.duracion = duracion
        self.calificacion = calificacion
        self.idioma = idioma
        self.estado = estado

# Matriz 3x3 de películas
coleccion = [[Pelicula() for _ in range(3)] for _ in range(3)]

def inicializarColeccion():
    coleccion[0][0] = Pelicula("El Conjuro", "James Wan", 2013, "Terror", 112, 8.0, "Ingles", "V")
    coleccion[0][1] = Pelicula("Hereditary", "Ari Aster", 2018, "Terror", 127, 8.5, "Ingles", "V")
    coleccion[0][2] = Pelicula("Rec", "Jaume Balaguero", 2007, "Terror", 78, 7.5, "Espanol", "N")

def mostrarColeccion():
    print("\n--- Colección de Películas de Terror ---")
    for fila in coleccion:
        for p in fila:
            if p.titulo:
                print(f"{p.titulo:15} | {p.director} | {p.anio} | {p.duracion} min | "
                      f"{p.calificacion}/10 | {p.idioma} | {p.estado}")

def buscarPorTitulo():
    titulo = input("Ingrese título: ")
    for fila in coleccion:
        for p in fila:
            if p.titulo == titulo:
                print(f"Encontrada: {p.titulo} ({p.anio}) - Dir: {p.director}")
                return
    print("Película no encontrada.")

def modificarPelicula():
    titulo = input("Ingrese título a modificar: ")
    for i in range(len(coleccion)):
        for j in range(len(coleccion[i])):
            if coleccion[i][j].titulo == titulo:
                coleccion[i][j].director = input("Nuevo director: ")
                coleccion[i][j].anio = int(input("Nuevo año: "))
                coleccion[i][j].duracion = float(input("Nueva duración: "))
                coleccion[i][j].calificacion = float(input("Nueva calificación: "))
                coleccion[i][j].idioma = input("Nuevo idioma: ")
                coleccion[i][j].estado = input("Nuevo estado (V/N): ")
                print("Modificada correctamente.")
                return
    print("Película no encontrada.")

def eliminarPelicula():
    titulo = input("Ingrese título a eliminar: ")
    for i in range(len(coleccion)):
        for j in range(len(coleccion[i])):
            if coleccion[i][j].titulo == titulo:
                coleccion[i][j] = Pelicula()
                print("Película eliminada.")
                return
    print("Película no encontrada.")

def promedioCalificacion():
    suma = 0
    contador = 0
    for fila in coleccion:
        for p in fila:
            if p.titulo:
                suma += p.calificacion
                contador += 1
    if contador > 0:
        print(f"Promedio: {suma/contador:.2f}/10")
    else:
        print("No hay películas.")

def menu():
    opcion = 0
    while opcion != 6:
        print("\n--- Menú ---")
        print("1. Mostrar")
        print("2. Buscar")
        print("3. Modificar")
        print("4. Eliminar")
        print("5. Promedio")
        print("6. Salir")
        opcion = int(input("Opción: "))

        if opcion == 1:
            mostrarColeccion()
        elif opcion == 2:
            buscarPorTitulo()
        elif opcion == 3:
            modificarPelicula()
        elif opcion == 4:
            eliminarPelicula()
        elif opcion == 5:
            promedioCalificacion()
        elif opcion == 6:
            print("Saliendo...")
        else:
            print("Opción inválida.")

# Programa principal
inicializarColeccion()
menu()
```
